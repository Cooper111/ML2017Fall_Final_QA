# shift ans to origin

```python
def shift_test_ans(test_ans_list, test_bias_list): 
    # assume test answer format : tuple ((start, end)) list 
    shifted_test_ans_list = []
    for ans_tuple, para_bias_list in zip(test_ans_list, test_bias_list):
        ans_start = ans_tuple[0]
        ans_end = ans_tuple[1]
        
        # shift
        ans_start += para_bias_list[ans_start]
        ans_end += para_bias_list[ans_end]
    
        shifted_test_ans_list.append((ans_start, ans_end))
```
